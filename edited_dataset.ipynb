```python
import pandas as pd
```

```python
labels_df = pd.read_csv('labels.csv')
data_df = pd.read_csv('data.csv')
```

```python
labels_df
```

```python
data_df
```

```python
olderName = data_df.columns[0]
newName = labels_df.columns[0]

data_df.rename(columns={olderName: newName}, inplace=True)

data_df
```

```python
# Merge the data based on a common identifier column
merged_df = pd.merge(labels_df, data_df, on='Sample')

merged_df
```

```python
merged_df.isnull().sum()
```

```python
all_zeros_columns = []

# Iterate over the columns of the DataFrame
for column in merged_df.columns:
    if (merged_df[column] == 0).all():
        all_zeros_columns.append(column)

# Print the column names with all zeros
if len(all_zeros_columns) > 0:
    print("Columns with all zeros:")
    for column_name in all_zeros_columns:
        print(column_name)
else:
    print("No columns have all zeros.")
```
